```python
import os
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from torch.utils.data import Dataset, DataLoader, ConcatDataset
from torchvision.datasets import VOCDetection
from torchvision.transforms import functional as F
from torchvision.models.detection import fcos_resnet50_fpn
from torchvision.models.detection.fcos import FCOS_ResNet50_FPN_Weights
from torchvision.utils import draw_bounding_boxes

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
print("Torch:", torch.__version__)
print("Torchvision:", torchvision.__version__)
```

```python
VOC_CLASSES = [
    "aeroplane", "bicycle", "bird", "boat", "bottle",
    "bus", "car", "cat", "chair", "cow",
    "diningtable", "dog", "horse", "motorbike", "person",
    "pottedplant", "sheep", "sofa", "train", "tvmonitor"
]

CLASS_TO_IDX = {cls: i + 1 for i, cls in enumerate(VOC_CLASSES)}
IDX_TO_CLASS = {i + 1: cls for i, cls in enumerate(VOC_CLASSES)}

print(CLASS_TO_IDX)
```

```python
class VOCDetectionTorchvision(Dataset):
    def __init__(self, year, image_set, root="./data", download=True):
        self.dataset = VOCDetection(
            root=root,
            year=year,
            image_set=image_set,
            download=download
        )

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, target = self.dataset[idx]

        objects = target["annotation"].get("object", [])
        if isinstance(objects, dict):
            objects = [objects]

        boxes = []
        labels = []

        for obj in objects:
            label_name = obj["name"]
            bbox = obj["bndbox"]

            xmin = float(bbox["xmin"])
            ymin = float(bbox["ymin"])
            xmax = float(bbox["xmax"])
            ymax = float(bbox["ymax"])

            if xmax > xmin and ymax > ymin:
                boxes.append([xmin, ymin, xmax, ymax])
                labels.append(CLASS_TO_IDX[label_name])

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)

        image_id = torch.tensor([idx])
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0]) if len(boxes) > 0 else torch.tensor([])
        iscrowd = torch.zeros((len(boxes),), dtype=torch.int64)

        img_tensor = F.to_tensor(img)

        target_dict = {
            "boxes": boxes,
            "labels": labels,
            "image_id": image_id,
            "area": area,
            "iscrowd": iscrowd
        }

        return img_tensor, target_dict
```

```python
voc2007_trainval = VOCDetectionTorchvision(year="2007", image_set="trainval")
voc2012_trainval = VOCDetectionTorchvision(year="2012", image_set="trainval")
voc2007_test = VOCDetectionTorchvision(year="2007", image_set="test")

train_dataset = ConcatDataset([voc2007_trainval, voc2012_trainval])
test_dataset = voc2007_test

print("VOC2007 trainval:", len(voc2007_trainval))
print("VOC2012 trainval:", len(voc2012_trainval))
print("Training set size:", len(train_dataset))
print("VOC2007 test size:", len(test_dataset))
```

```python
def collate_fn(batch):
    return tuple(zip(*batch))

train_loader = DataLoader(
    train_dataset,
    batch_size=2,
    shuffle=True,
    num_workers=2,
    collate_fn=collate_fn
)

test_loader = DataLoader(
    test_dataset,
    batch_size=1,
    shuffle=False,
    num_workers=2,
    collate_fn=collate_fn
)
```

```python
weights = FCOS_ResNet50_FPN_Weights.DEFAULT

model = fcos_resnet50_fpn(
    weights=None,
    num_classes=21
)

model.to(device)

optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.005,
    momentum=0.9,
    weight_decay=0.0001
)

print("Model loaded.")
```

```python
max_iters = 1000
print_interval = 50

model.train()
iteration = 0

for images, targets in train_loader:
    images = [img.to(device) for img in images]
    targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

    loss_dict = model(images, targets)
    losses = sum(loss for loss in loss_dict.values())

    optimizer.zero_grad()
    losses.backward()
    optimizer.step()

    if iteration % print_interval == 0:
        print(
            f"Iter {iteration}/{max_iters} | "
            f"Loss: {losses.item():.4f} | "
            + " | ".join([f"{k}: {v.item():.4f}" for k, v in loss_dict.items()])
        )

    iteration += 1

    if iteration >= max_iters:
        break

print("Training finished.")
```

```python
def box_iou(boxes1, boxes2):
    if len(boxes1) == 0 or len(boxes2) == 0:
        return torch.zeros((len(boxes1), len(boxes2)))

    area1 = (boxes1[:, 2] - boxes1[:, 0]).clamp(min=0) * (boxes1[:, 3] - boxes1[:, 1]).clamp(min=0)
    area2 = (boxes2[:, 2] - boxes2[:, 0]).clamp(min=0) * (boxes2[:, 3] - boxes2[:, 1]).clamp(min=0)

    lt = torch.max(boxes1[:, None, :2], boxes2[:, :2])
    rb = torch.min(boxes1[:, None, 2:], boxes2[:, 2:])

    wh = (rb - lt).clamp(min=0)
    inter = wh[:, :, 0] * wh[:, :, 1]

    union = area1[:, None] + area2 - inter
    return inter / union.clamp(min=1e-6)


def compute_ap_11point(recalls, precisions):
    ap = 0.0
    for t in np.arange(0.0, 1.1, 0.1):
        if np.sum(recalls >= t) == 0:
            p = 0
        else:
            p = np.max(precisions[recalls >= t])
        ap += p / 11.0
    return ap


def evaluate_map(model, data_loader, device, iou_threshold=0.5, score_threshold=0.05, max_images=500):
    model.eval()

    all_predictions = {cls_idx: [] for cls_idx in range(1, 21)}
    all_gts = {cls_idx: {} for cls_idx in range(1, 21)}

    image_counter = 0

    with torch.no_grad():
        for images, targets in data_loader:
            images_gpu = [img.to(device) for img in images]
            outputs = model(images_gpu)

            for img, target, output in zip(images, targets, outputs):
                image_id = image_counter

                gt_boxes = target["boxes"].cpu()
                gt_labels = target["labels"].cpu()

                for cls_idx in range(1, 21):
                    cls_gt_boxes = gt_boxes[gt_labels == cls_idx]
                    all_gts[cls_idx][image_id] = {
                        "boxes": cls_gt_boxes,
                        "detected": torch.zeros(len(cls_gt_boxes), dtype=torch.bool)
                    }

                pred_boxes = output["boxes"].cpu()
                pred_scores = output["scores"].cpu()
                pred_labels = output["labels"].cpu()

                keep = pred_scores >= score_threshold
                pred_boxes = pred_boxes[keep]
                pred_scores = pred_scores[keep]
                pred_labels = pred_labels[keep]

                for box, score, label in zip(pred_boxes, pred_scores, pred_labels):
                    label = int(label.item())

                    # COCO pretrained labels do not perfectly match VOC labels.
                    # We only evaluate label IDs 1..20 for a simple VOC-style alternative metric.
                    if 1 <= label <= 20:
                        all_predictions[label].append({
                            "image_id": image_id,
                            "box": box,
                            "score": float(score.item())
                        })

                image_counter += 1
                if image_counter >= max_images:
                    break

            if image_counter >= max_images:
                break

    aps = {}

    for cls_idx in range(1, 21):
        preds = all_predictions[cls_idx]
        preds = sorted(preds, key=lambda x: x["score"], reverse=True)

        total_gts = sum(len(v["boxes"]) for v in all_gts[cls_idx].values())

        if total_gts == 0:
            aps[cls_idx] = np.nan
            continue

        tp = np.zeros(len(preds))
        fp = np.zeros(len(preds))

        for i, pred in enumerate(preds):
            image_id = pred["image_id"]
            pred_box = pred["box"].unsqueeze(0)

            gt_info = all_gts[cls_idx].get(image_id, {"boxes": torch.empty((0, 4)), "detected": torch.empty((0,), dtype=torch.bool)})
            gt_boxes = gt_info["boxes"]

            if len(gt_boxes) == 0:
                fp[i] = 1
                continue

            ious = box_iou(pred_box, gt_boxes).squeeze(0)
            max_iou, max_idx = torch.max(ious, dim=0)

            if max_iou >= iou_threshold and not gt_info["detected"][max_idx]:
                tp[i] = 1
                gt_info["detected"][max_idx] = True
            else:
                fp[i] = 1

        tp_cumsum = np.cumsum(tp)
        fp_cumsum = np.cumsum(fp)

        recalls = tp_cumsum / max(total_gts, 1)
        precisions = tp_cumsum / np.maximum(tp_cumsum + fp_cumsum, 1e-6)

        aps[cls_idx] = compute_ap_11point(recalls, precisions)

    valid_aps = [ap for ap in aps.values() if not np.isnan(ap)]
    mean_ap = float(np.mean(valid_aps)) if valid_aps else 0.0

    return mean_ap, aps
```

```python
mAP, aps = evaluate_map(
    model,
    test_loader,
    device,
    iou_threshold=0.5,
    score_threshold=0.05,
    max_images=500
)

print("VOC-style mAP@0.5:", mAP)

for cls_idx, ap in aps.items():
    print(f"{IDX_TO_CLASS[cls_idx]}: {ap:.4f}")
```

```python
def visualize_prediction(model, dataset, index=0, score_threshold=0.5):
    model.eval()

    image, target = dataset[index]
    image_gpu = image.to(device)

    with torch.no_grad():
        output = model([image_gpu])[0]

    boxes = output["boxes"].detach().cpu()
    scores = output["scores"].detach().cpu()
    labels = output["labels"].detach().cpu()

    keep = scores >= score_threshold
    boxes = boxes[keep]
    scores = scores[keep]
    labels = labels[keep]

    img_uint8 = (image * 255).byte()

    label_texts = []
    for label, score in zip(labels, scores):
        label_id = int(label.item())
        name = IDX_TO_CLASS.get(label_id, f"class_{label_id}")
        label_texts.append(f"{name}: {score:.2f}")

    drawn = draw_bounding_boxes(
        img_uint8,
        boxes,
        labels=label_texts,
        width=2
    )

    plt.figure(figsize=(10, 8))
    plt.imshow(drawn.permute(1, 2, 0))
    plt.axis("off")
    plt.show()

visualize_prediction(model, test_dataset, index=10, score_threshold=0.5)
```
